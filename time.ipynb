```python
import seaborn as sns
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
from bs4 import BeautifulSoup
import glob, csv
file_names = glob.glob("messages/*.html")
message_dates = []
for file_name in file_names:
    f = open(file_name, 'r').read()
    soup = BeautifulSoup(f, 'lxml')
    times = soup.findAll("span", { "class" : "meta" })
    times = [str(i)[19:-7] for i in times]
    message_dates.append(times)
```

```python
message_times = []
for convo in message_dates:
    for i in convo:
        message_times.append(i[message_dates[1][0].find('at')+ 3:34])
```

```python
len(message_times)
```

```python
message_times[1][4]
```

```python
message_hours = []
for i in message_times:
    if i[4] == 'p':
        message_hours.append(int(i[0]) + 12)
    elif i[4] == 'a':
        message_hours.append(int(i[0]))
```

```python
sns.countplot(message_hours)
```
